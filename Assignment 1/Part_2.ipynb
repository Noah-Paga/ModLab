
## Limiting behaviour of the Langmuir isotherm

The Langmuir isotherm is given by  

$$
q(P,T) = \frac{q_{\text{sat}} \, k(T) \, P}{1 + k(T) \, P}
$$

where  
* $q$ is the uptake (mol·kg⁻¹),  
* $q_{\text{sat}}$ is the saturation capacity (mol·kg⁻¹),  
* $k(T)$ is the equilibrium constant (Pa⁻¹),  
* $P$ is the pressure (Pa).

---

### (a) High‑pressure limit $P \to \infty$ — uptake reaches plateau

Divide numerator and denominator by $k(T)P$:
$$
q = \frac{q_{\text{sat}}}{\dfrac{1}{k(T)P} + 1}.
$$

As $P \to \infty$, the term $\dfrac{1}{k(T)P} \to 0$. Therefore
$$
\boxed{ q \xrightarrow[P\to\infty]{} q_{\text{sat}} }.
$$

**Interpretation:** at very large pressure all adsorption sites are occupied, so the uptake approaches the saturation loading $q_{\text{sat}}$.

---

### (b) Low‑pressure limit $P \to 0$ — linear dependence on $P$

For small pressures $k(T)P \ll 1$.  
Approximate the denominator:
$$
q \approx \frac{q_{\text{sat}} k(T) P}{1} = q_{\text{sat}} k(T) P.
$$

Thus
$$
\boxed{ q \approx q_{\text{sat}} k(T) P \quad (P \to 0) }.
$$

This is linear in $P$ with slope $q_{\text{sat}} k(T)$.
