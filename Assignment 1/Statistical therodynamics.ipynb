# Part B: Statistical Thermodynamics

### 1) Canonical Ensemble

The **canonical ensemble** describes a system with  
**fixed number of particles (N)**, **fixed volume (V)**, and **fixed temperature (T)**.  

The system can be in many **microstates**, each with its own **energy** $E_v$. The **probability** of a microstate is given by the **Boltzmann factor**:

<p align="center">

$$
P_v = \frac{e^{-E_v / (k_B T)}}{Q}
$$

</p>

where $k_B$ is **Boltzmann’s constant**, $T$ is the **temperature**, and $Q$ is the **partition function**.  


For **adsorption**, this means gas molecules in a porous solid at constant **T** and **V**. Each way the molecules sit on the surface is a **microstate**. The canonical ensemble explains how likely each configuration is, and allows us to calculate **uptake** and **heat of adsorption**.

### 2) Partition Function (Q)

In the **canonical ensemble**, the **partition function** is the sum of all **Boltzmann factors** of the possible **microstates**:

$$
Q(N,V,T) = \sum_{v} e^{-E_v / (k_B T)}
$$

Here $E_v$ is the **energy** of microstate $v$, $k_B$ is **Boltzmann’s constant**, and $T$ is the **temperature**.  

For **adsorption**, $Q$ counts all possible **arrangements** of molecules in the adsorbent.  
It is the key quantity to calculate **probabilities** and link the microscopic description to **macroscopic properties**.

### 3) Probability of a Microstate

The **probability** of finding the system in a given **microstate** $v$ with energy $E_v$ is:

$$
P_v = \frac{e^{-E_v / (k_B T)}}{Q}
$$

Here $Q$ is the **partition function**, $k_B$ is **Boltzmann’s constant**, and $T$ is the **temperature**.  

In **adsorption**, this means the chance of a certain arrangement of molecules on the surface.  
Microstates with **lower energy** are more probable, but all microstates contribute through the Boltzmann factor.  
This is how statistical mechanics connects the **microscopic states** to the **macroscopic behavior** of adsorption.

### 4) Thermodynamic Relations and Full Derivations

We use the canonical ensemble (N, V, T) with

$$
Q(N,V,T)=\sum_{v} e^{-E_v/(k_B T)},\qquad
P_v=\frac{e^{-E_v/(k_B T)}}{Q},\qquad
\beta=\frac{1}{k_B T}.
$$



#### (a) Entropy

Goal
$$
\langle S\rangle = k_B \ln Q + k_B T\left(\frac{\partial \ln Q}{\partial T}\right)_{N,V}.
$$

Step 1: Definition
$$
S=-k_B\sum_v P_v \ln P_v.
$$

Step 2: Insert $P_v$
Since $P_v = \dfrac{e^{-\beta E_v}}{Q}$, we have $\ln P_v = -\beta E_v - \ln Q$. Then

$$
S = -k_B \sum_v P_v(-\beta E_v - \ln Q)
= k_B \beta \sum_v P_v E_v + k_B \ln Q \sum_v P_v.
$$

Step 3: Simplify
$$
S = k_B \beta \langle E \rangle + k_B \ln Q.
$$

Step 4: Express $\langle E \rangle$
$$
\frac{\partial \ln Q}{\partial T}
= \frac{1}{Q}\sum_v e^{-E_v/(k_B T)}\frac{E_v}{k_B T^2}
= \frac{\langle E\rangle}{k_B T^2}.
$$

Step 5: Finish
$$
S = k_B \ln Q + k_B T \left(\frac{\partial \ln Q}{\partial T}\right)_{N,V}.
$$




#### (b) Pressure

Goal
$$
P = k_B T \left(\frac{\partial \ln Q}{\partial V}\right)_{N,T}.
$$

Step 1: Helmholtz free energy
$$
F=-k_B T \ln Q.
$$

Step 2: Definition of pressure
$$
P = -\left(\frac{\partial F}{\partial V}\right)_{N,T}.
$$

Step 3: Differentiate
$$
\left(\frac{\partial F}{\partial V}\right){N,T}
= -,k_B T \left(\frac{\partial \ln Q}{\partial V}\right){N,T}.
$$

Step 4: Combine
$$
P = k_B T \left(\frac{\partial \ln Q}{\partial V}\right)_{N,T}.
$$




#### (c) Chemical potential

Goal
$$
\mu = -,k_B T \left(\frac{\partial \ln Q}{\partial N}\right)_{V,T}.
$$

Step 1: Free energy
$$
F=-k_B T \ln Q.
$$

Step 2: Definition of chemical potential
$$
\mu=\left(\frac{\partial F}{\partial N}\right)_{V,T}.
$$

Step 3: Differentiate
$$
\mu = -,k_B T \left(\frac{\partial \ln Q}{\partial N}\right)_{V,T}.
$$




#### (d) Internal energy

Goal
$$
\langle E\rangle = k_B T^2 \left(\frac{\partial \ln Q}{\partial T}\right)_{V,N}.
$$

Method A: From $\partial_T \ln Q$
From (a) Step 4 we have
$$
\frac{\partial \ln Q}{\partial T} = \frac{\langle E\rangle}{k_B T^2}.
$$
Rearrange:
$$
\langle E\rangle = k_B T^2 \left(\frac{\partial \ln Q}{\partial T}\right)_{V,N}.
$$

Method B: From $\partial_\beta \ln Q$
$$
\frac{\partial \ln Q}{\partial \beta}
= \frac{1}{Q}\sum_v (-E_v)e^{-\beta E_v}
= -\langle E\rangle,
$$
and with $\dfrac{\partial \beta}{\partial T}=-\dfrac{1}{k_B T^2}$,
$$
\frac{\partial \ln Q}{\partial T}
=\frac{\partial \ln Q}{\partial \beta}\frac{\partial \beta}{\partial T}
=(-\langle E\rangle)\left(-\frac{1}{k_B T^2}\right)
=\frac{\langle E\rangle}{k_B T^2}.
$$

So,
$$
\langle E\rangle = k_B T^2 \left(\frac{\partial \ln Q}{\partial T}\right)_{V,N}.
$$
