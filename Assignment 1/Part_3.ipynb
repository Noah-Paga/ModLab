## 3. Temperature dependence of Henry’s coefficient

**Definition:**  
Henry’s coefficient \(K_H(T)\) is the initial slope of the isotherm at \(P \to 0\):
\[
K_H(T) \equiv \left. \frac{\partial q}{\partial P} \right|_{P \to 0} = q_{\text{sat}}\,k(T).
\]

**Van’t Hoff form for \(k(T)\):**  
\[
k(T) = k_0 \exp\!\left(-\frac{\Delta H}{R T}\right),
\]
so
\[
K_H(T) = q_{\text{sat}}\,k_0 \exp\!\left(-\frac{\Delta H}{R T}\right).
\]

**Interpretation:**  
\(K_H\) follows an Arrhenius-like (exponential) dependence on \(1/T\).  
If \(\Delta H < 0\) (exothermic adsorption), \(K_H\) decreases as temperature \(T\) increases.
